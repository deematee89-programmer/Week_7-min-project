```python
"""
Day 35 Activity: EDA Mini-Project
Tasks:
1) Run univariate, bivariate, and correlation analysis
2) Use at least 5 plots
3) Write 3 insights
"""

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# TODO: Load data from data/day35_project.csv
# df = pd.read_csv(...)

# TODO: Perform EDA and write insights


path="/kaggle/input/datasets/deemass/min-project-w7/day35_project.csv"
df=pd.read_csv(path)

#==============  univariate & bivariate analysis===========
numeric_cols = df.select_dtypes(include=["number"]).columns
cat_cols = df.select_dtypes(include=["object","category"]).columns

for col in numeric_cols[:4]:
    print("*"*100)
    print(f"Column: {col}")
    print("*"*100)
    print("Univariate analysis")
    print("="*100)
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.show()
    print("Bivariate analysis")
    print("="*100)
    sns.boxplot(x=cat_cols[0], y=col, data=df)
    plt.show()
#============ correlation analysis=============
print("*"*100)
print("Correlation analysis")
print("*"*100)
sns.heatmap(df[numeric_cols].corr(), cmap="coolwarm", vmin=-1, vmax=1)
g = sns.relplot(data=df, x=numeric_cols[0], y=numeric_cols[1],hue=cat_cols[0], col=cat_cols[0], kind="scatter")
```
